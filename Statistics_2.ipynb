```python
mean_temperature=np.mean(ds['Temperature (C)'])
print("The mean temperature is", mean_temperature)
```

```python
varience=np.var(ds['Temperature (C)'])
print("The varience temperature is", varience)
```

```python
st=np.std(ds['Temperature (C)'])
print("Standard Deviation Temp is", st)
```

```python
for i in range(1,13):
  month=ds.loc[ds['month']==i]['Temperature (C)']
  print("For Month "+str(i))
  print("Mean Temperature:"+str(np.mean(month)))
  print("Standard Deviation temperature:"+str(np.std(month)))
```
